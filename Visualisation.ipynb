<font size = 6><b><u> Visualising Data</u></b></font>

<font size =4><b>The follwing code uses seaborn to visualise the dataset and how the features vary with each other. </b></font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pd.read_csv('Dataset/P1.csv')
df
```

<font size = 4><b>Visualising the trend between IRRADIATION and DC_POWER using scatter plot</b></font>

```python
sn.regplot(x = "IRRADIATION", y="DC_POWER", data=df, fit_reg = True, scatter_kws={"alpha": 0.2})
```

<font size = 4><b>Visualising the trend between AMBIENT_TEMPERATURE and DC_POWER using scatter plot</b></font>

```python
sn.regplot(x = "AMBIENT_TEMPERATURE", y="DC_POWER", data=df, fit_reg = True, scatter_kws={"alpha": 0.2})
```

<font size = 4><b>Visualising the trend between MODULE_TEMPERATURE and DC_POWER using scatter plot</b></font>

```python
sn.regplot(x = "MODULE_TEMPERATURE", y="DC_POWER", data=df, fit_reg = True, scatter_kws={"alpha": 0.2})
```

<font size = 4><b>Visualising the correlation between the four variables (consider the column 'Unamed: 0' null as it is used for indexing the dataset).</b></font>

```python
columns=[]
corr = df.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(df.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
plt.show()
```
